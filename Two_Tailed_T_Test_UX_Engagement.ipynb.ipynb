```python
import numpy as np 
import scipy.stats as stat
from scipy import stats
import seaborn as sns 
import matplotlib.pyplot as plt

np.random.seed(42) 
design_A = np.random.normal(loc=45,scale=10,size=100) 
design_B = np.random.normal(loc=52,scale=12,size=100) 
print(f"Design-A {design_A}") 
print(f"Design-B {design_B}") 
```

```python
n1,n2 = len(design_A),len(design_B) 
mean_A,mean_B = np.mean(design_A),np.mean(design_B) 
std_A,std_B = np.std(design_A),np.std(design_B) 
df = n1 + n2 - 2 
critical_val = stats.t.ppf(1-0.05/2,df) 
t_cal = (mean_A-mean_B)/np.sqrt((std_A/n1)+(std_B/n2)) 
print(f"Crititcal-Value (Both Sides): {critical_val:4f}") 
print(f"T-Calculated-Value: {t_cal:4f}") 
if abs(t_cal) > critical_val: 
    print("Decision: Hypothesis Rejected") 
else: 
    print("Decision: Hypothesis not Rejected") 
```

```python
plt.figure(figsize=(10,6)) 
sns.kdeplot(design_A,fill=True,color="#3498db",label="Design A (User Time)",alpha=0.4) 
sns.kdeplot(design_B,fill=True,color="#e67e22",label="Design B (User Time)",alpha=0.4) 
plt.title("Two-Tailed T-Test: Website Design A Design B User Engagement",fontsize=11,fontweight='bold') 
plt.xlabel("Time Spend on Website (Seconds)",fontsize=11) 
plt.ylabel("Density",fontsize=11)
plt.legend(loc="upper right") 
plt.tight_layout()
plt.show()
```
